# Downloading and preparing archived data

A set of archived data are available at [TODO: insert link/DOI]. This dataset allows you to:
- Use existing unseen-awg simulations.
- Generate new simulations.
- Create new unseen-awg instances.

The archived data consist of NetCDF4 files and Zarr stores. For practical reasons, the Zarr stores are archived as multi-part (split) zip files. When using the Zarr stores, you must unzip all partial files for a given dataset together; otherwise, the resulting store will suffer from missing data.

(extracting-data)=
## Extracting the data

We provide a script called unzip_provided_datasets.sh to facilitate a safe extraction. The general syntax is:
```Bash
./unzip_provided_datasets.sh <dir_containing_zipped_files> <dir_target>
```

## Snakemake workflow integration

In `workflow/`, we also provide a Snakemake workflow to facilitate using this data. The extraction script automatically generates directory and file names that are compatible with this workflow (e.g., it includes parameter hashes in the file names).

To use the workflow, the unzipped data must be placed in the specific directories defined in `configs/paths.yaml` (the paths can be modified in this file).

### 1. Extract the input datasets:

```Bash
./unzip_provided_datasets.sh dir_with_partial_files dir_preprocessed_datasets
```

####2. Extract the weather generator instances:

```Bash
./unzip_provided_datasets.sh dir_with_partial_files dir_wgs
````

After extraction, your resulting directory paths should look like this:

    dir_wgs/wg_reforecasts_5e06172f_f40e9460_1e69bda9
    dir_preprocessed_datasets/preprocessed_impact_variables_reforecasts/combined_7d1d3d97.zarr
    dir_preprocessed_datasets/preprocessed_impact_variables_era5/combined_facc0e91.zarr
    dir_preprocessed_datasets/preprocessed_circulation_reforecasts/combined_5e06172f.zarr
    dir_preprocessed_datasets/preprocessed_circulation_era5/combined_f3d1f2f7.zarr
